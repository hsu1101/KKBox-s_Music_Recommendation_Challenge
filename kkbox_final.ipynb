### <font color =#000000>Object to Category </font>

```python
def obtocat(df):
    object_cols = list(df.select_dtypes(include=['object']).columns)
    for col in object_cols:
        df[col]=df[col].astype('category')
```

```python
obtocat(train)
obtocat(test)
obtocat(songs)
obtocat(members)
```

### <font color =#000000>Isrc to Year and then drop name and isrc columns </font>

```python
def isrctoyear(isrc):
    if type(isrc) == str:
        if int(isrc[5:7]) > 17:
            return 1900 + int(isrc[5:7])
        else:
            return 2000 + int(isrc[5:7])
    else:
        return np.nan
```

```python
songs_extra['song_year'] = songs_extra['isrc'].apply(isrctoyear)
```

```python
songs_extra.drop(['isrc', 'name'], axis = 1, inplace = True)
```

### <font color =#000000>Add new columns(genre_id_count) </font>

```python
def genreidcount(x):
    if x == 'no_genre_id':
        return 0
    else:
        return x.count('|') + 1
```

```python
train['genre_ids'].cat.add_categories('no_genre_id',inplace = True)
test['genre_ids'].cat.add_categories('no_genre_id',inplace=True)
```

```python
train['genre_ids'].fillna('no_genre_id',inplace=True)
test['genre_ids'].fillna('no_genre_id',inplace=True)
```

```python
train['genre_id_count'] = train['genre_ids'].apply(genreidcount).astype(np.int8)
test['genre_id_count'] = test['genre_ids'].apply(genreidcount).astype(np.int8)
```

### <font color =#000000>Add new columns(lyricists_count) </font>

```python
def lyricistcount(x):
    if x == 'no_lyricist':
        return 0
    else:
        return sum(map(x.count, ['|', '/', '\\', ';'])) + 1
    return sum(map(x.count, ['|', '/', '\\', ';']))
```

```python
train['lyricist'].cat.add_categories('no_lyricist',inplace = True)
test['lyricist'].cat.add_categories('no_lyricist',inplace=True)
```

```python
train['lyricist'].fillna('no_lyricist',inplace=True)
test['lyricist'].fillna('no_lyricist',inplace=True)
```

```python
train['lyricists_count'] = train['lyricist'].apply(lyricistcount).astype(np.int8)
test['lyricists_count'] = test['lyricist'].apply(lyricistcount).astype(np.int8)
```

### <font color =#000000>Add new columns(artist_count) </font>

```python
def artistcount(x):
    if x == 'no_artist':
        return 0
    else:
        return x.count('and') + x.count(',') + x.count('feat') + x.count('&')
```

```python
train['artist_name'].cat.add_categories('no_artist',inplace = True)
test['artist_name'].cat.add_categories('no_artist',inplace=True)
```

```python
train['artist_name'].fillna('no_artist',inplace=True)
test['artist_name'].fillna('no_artist',inplace=True)
```

```python
train['artist_count'] = train['artist_name'].apply(artistcount).astype(np.int8)
test['artist_count'] = test['artist_name'].apply(artistcount).astype(np.int8)
```

### <font color =#000000>Import </font>

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

### <font color =#000000>load dataset </font>

```python
after_train = pd.read_csv('after_train.csv')
after_test = pd.read_csv('after_test.csv')
```

### <font color =#000000>Drop columns </font>

```python
after_train = after_train.drop(after_train.columns[0], axis=1)
after_test = after_test.drop(after_test.columns[0], axis=1)
```

```python
after_train[:5]
```

```python
after_test = after_test.drop(columns='id',axis=1)
```

```python
after_test[:5]
```

### <font color =#000000>Fillna 'NaN' </font>

```python
after_train = after_train.fillna('NaN')
after_test = after_test.fillna('NaN')
```

### <font color =#000000>Encoded data(to number) </font>

```python
from sklearn import preprocessing

def encoder(after_train):
    le = preprocessing.LabelEncoder()
    for column_name in after_train.columns:
            if after_train[column_name].dtype.name in ['category','object']:
                after_train[column_name] = le.fit_transform(after_train[column_name].astype(str))
```

```python
encoder(after_train)
encoder(after_test)
```

### <font color =#000000> after_train to X y</font>

```python
X = after_train.drop(columns='target',axis=1)
y = after_train['target']
```

### <font color =#000000> train test split  </font>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4,random_state=0)
```

### <font color =#000000>Random select 500000 to train </font>

### <font color =#000000>LogisticRegression </font>

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train_sampled,y_train_sampled)
```

### <font color =#000000>LogisticRegression Result </font>

```python
print(lr.score(X_train_sampled,y_train_sampled))
print(lr.score(X_train,y_train))
print(lr.score(X_test,y_test))
```
